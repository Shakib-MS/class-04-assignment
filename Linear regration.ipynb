# Separate Dependent and Independent varriable

```python
df.head()
```

```python
x=df[['Height']]
y=df['Weight']
```

```python
x.head()
```

```python
y.head()
```

```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
#train=60%
#test=40%
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.40)
```

```python
xtest.head()
```

```python
xtrain.head()
```

```python
xtrain.shape#60%
```

```python
xtest.shape #40%
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
reg.fit(xtrain,ytrain) #training model
```

```python
reg.predict([[73.847017]])
```

```python
reg.predict([[71.730978]])
```

```python
ypred=reg.predict(xtest)
ypred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(ytest,ypred)
```

```python
mse
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mea=mean_absolute_error(ytest,ypred)
```
